```python
import pandas as pd
a=pd.read_csv('/content/task2.csv')
a
a.head()
```

```python
# data cleaning
# 1 Check for Missing Values
a.isnull().sum()
a
```

```python

# To replace missing values with mean
a['bmi'].fillna(a['bmi'].mean(), inplace=True)
a
```

```python
#2  Convert Categorical Variables to Numeric we use one-hot encoding
b = pd.get_dummies(a)
b
```

```python
#Exploratory Data Analysis (EDA)
# 1 summary statistics
a.describe()
```

```python
#2 data viualization :
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.barplot(x='gender',y='age',hue='stroke',data=a,errorbar=None)
plt.title("Average age by Gender and Stroke Status")
plt.xlabel('Gender')
plt.ylabel("Average Age")
plt.show()
```

The bar chart shows the average age by gender and stroke status.For males, the average age of those who have not had a stroke (blue) is around 40 years, while those who have had a stroke (orange) have an average age of around 70 years.
For females, the average age of those who have not had a stroke (blue) is also around 40 years, and the average age of those who have had a stroke (orange) is around 70 years.
For the 'Other' gender category, only the average age of those who have not had a stroke (blue) is shown, which is around 25 years. There is no data shown for those who have had a stroke.There is a clear difference in average age between individuals who have had a stroke and those who have not. For both males and females, individuals who have had a stroke are significantly older (around 70 years) compared to those who have not had a stroke (around 40 years).

```python
# 3 Correlation Heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(b.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 4 boxplot
plt.figure(figsize=(5,5))
sns.boxplot(a.bmi)
plt.title('bmi')
plt.show()
```

The median BMI appears to be around 30.
The IQR for BMI spans from about 25 to 35.
The lower whisker extends to around 15, and the upper whisker extends to around 45.
There are several outliers above the upper whisker, indicating a number of individuals with a BMI significantly higher than the majority. Some outliers reach BMI values as high as 100.

```python
# scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(a.age	,color='black')
plt.title('age')
plt.show()
```

In the context of heart stroke data, this scatter plot indicates that the dataset covers a wide age range, which is crucial for understanding how age influences stroke risk. This comprehensive age distribution allows for robust statistical analysis and ensures that conclusions drawn about age-related stroke risk factors are based on a diverse sample of individuals.A dense concentration of data points is observed across the age range, with no significant gaps.
The plot shows a consistent number of individuals across different ages, which may imply a well-represented age distribution in the dataset.
